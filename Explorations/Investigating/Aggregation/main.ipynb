```python

# Load the dataset
df = pd.read_csv("trend_data.csv")

# Check the first few rows
print(df.head())

# Summary statistics by Grouping
group_summary = df.groupby("Grouping")[["DL", "Consum"]].describe()
print(group_summary)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv("trend_data.csv")

# Convert Month to datetime for better analysis
df["Month"] = pd.to_datetime(df["Month"])

# Summary statistics by Grouping
group_summary = df.groupby("Grouping")[["DL", "Consum"]].describe()
print("Summary Statistics:\n", group_summary)

# Plot trends for each grouping
plt.figure(figsize=(12, 6))
for group in df["Grouping"].unique():
    subset = df[df["Grouping"] == group]
    plt.plot(subset["Month"], subset["DL"], label=f"DL - {group}")
    plt.plot(subset["Month"], subset["Consum"], linestyle='dashed', label=f"Consum - {group}")
plt.legend()
plt.title("Trends of DL and Consum Over Time")
plt.xlabel("Month")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.show()

# Correlation Analysis
correlations = df.groupby("Grouping")[["DL", "Consum"]].corr().iloc[0::2, -1]
print("\nCorrelation between DL and Consum for each Grouping:\n", correlations)

# Detect seasonality using autocorrelation
from statsmodels.graphics.tsaplots import plot_acf
plt.figure(figsize=(12, 6))
for group in df["Grouping"].unique():
    subset = df[df["Grouping"] == group]
    plt.subplot(2, 3, list(df["Grouping"].unique()).index(group) + 1)
    plot_acf(subset["DL"], lags=6, title=f"ACF - {group}", ax=plt.gca())
plt.tight_layout()
plt.show()

# Outlier Detection using Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(x="Grouping", y="DL", data=df)
plt.title("Outlier Detection in DL")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x="Grouping", y="Consum", data=df)
plt.title("Outlier Detection in Consum")
plt.show()

# Rate of Change (Month-over-Month Percentage Change)
df["DL_Change"] = df.groupby("Grouping")["DL"].pct_change() * 100
df["Consum_Change"] = df.groupby("Grouping")["Consum"].pct_change() * 100
print("\nMonth-over-Month Change Summary:\n", df.groupby("Grouping")[["DL_Change", "Consum_Change"]].describe())
```
